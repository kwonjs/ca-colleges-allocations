```python
ca_pub_col = pd.read_csv('CA-public-colleges.csv')
ca_pub_col
```

```python
# do left or right join because we want to prioritize the allocations dataset and get 
# the locations of every university/campus on the ALLOCATIONS dataset 
```

This dataset is **California public universities & colleges** from these [search results]("https://nces.ed.gov/collegenavigator/?s=CA&ct=1&pg=1") from College Navigator

```python
ca_pub_col.drop(ca_pub_col.tail(3).index, inplace = True) 
ca_pub_col
```

```python
allocations
```

```python
ca_pub_col['IPEDS ID'] = ca_pub_col['IPEDS ID'].map(lambda x : int(x), na_action='ignore')
ca_pub_col['OPE ID'] = ca_pub_col['OPE ID'].map(lambda x : int(x), na_action='ignore')
allocations_tofloat = allocations.copy()
allocations_tofloat["OPEID"] = allocations_tofloat['OPEID'].map(lambda x : float(x))
allocations_tofloat #god i can't believe i have to do this this is disgusting
allocations_tofloat = allocations_tofloat.rename(columns={"OPEID": "OPE ID"})
allocations_tofloat
```

```python
ca_pub_merged = ca_pub_col.merge(allocations_tofloat, how = 'left', on = 'OPE ID')
ca_pub_merged
```

```python
ca_pub_merged['OPE ID'] = ca_pub_merged['OPE ID'].map(lambda x : int(x), na_action='ignore')
ca_pub_merged
```

```python
# merge colnav_public + colnav_privnonprofit + colnav_privforprofit sheets
```

```python
ca_pub_col
```

```python
ca_privnonprofit_col = pd.read_csv('colnav_privnonprof.csv')
```

```python
ca_privforprofit_col = pd.read_csv('colnav_privforprof.csv')
```

```python
ca_privnonprofit_col.drop(ca_privnonprofit_col.tail(3).index, inplace = True)
ca_privforprofit_col.drop(ca_privforprofit_col.tail(3).index, inplace = True)
```

```python
ca_privforprofit_col
```

```python
total_ca_col = [ca_pub_col, ca_privnonprofit_col, ca_privforprofit_col]
combined_ca_colleges = pd.concat(total_ca_col)
combined_ca_colleges.describe
```

```python
allocations_for_ca_colleges = allocations_tofloat.merge(combined_ca_colleges, how = 'left', on = 'OPE ID')
```

```python
allocations_for_ca_colleges
```

```python
# drop NaN values in "Address" column of combined dataset

allocations_for_ca_colleges_dropped = allocations_for_ca_colleges[pd.notnull(allocations_for_ca_colleges['Address'])]
```

```python
allocations_for_ca_colleges_dropped.columns
```

```python
columns_to_keep = ['OPE ID', 'School', 'Total Allocation ($)',
       'Minimum Allocation to\nbe Awarded for\nEmergency Financial Aid\nGrants to Students ($)',
                  'Address', 'Type', 'Student population',
       'Undergraduate students', 'Graduation Rate', 'Transfer-Out Rate']
allocations_for_ca_colleges_dropped_cleaned = allocations_for_ca_colleges_dropped[columns_to_keep]
allocations_for_ca_colleges_dropped_cleaned
```

```python
allocations_for_ca_colleges_dropped_cleaned.sort_values(by=['Student population'], ascending = False)
```

```python
allocations_for_ca_colleges_dropped_cleaned['OPE ID'] = allocations_for_ca_colleges_dropped_cleaned['OPE ID'].map(lambda x : int(x))
allocations_for_ca_colleges_dropped_cleaned = allocations_for_ca_colleges_dropped_cleaned.rename(columns={"Minimum Allocation to\nbe Awarded for\nEmergency Financial Aid\nGrants to Students ($)": "Minimum Allocation to be Awarded for Emergency Financial Aid Grants to Students ($)"})
```

```python
allocations_for_ca_colleges_dropped_cleaned
```

```python
allocations_for_ca_colleges_dropped_cleaned.to_csv('allocations_ca_colleges.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

tot_allocation_y = allocations_for_ca_colleges_dropped_cleaned["Total Allocation ($)"]
student_pop_x = allocations_for_ca_colleges_dropped_cleaned["Student population"]

plt.plot(student_pop_x, tot_allocation_y)
```
